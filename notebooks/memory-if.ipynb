```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../memory_data.csv", delimiter=',', dtype=float)

capacities = data[:, 0]
speeds = data[:, 1]
prices = data[:, 2]

price_per_mb = prices / capacities
data = np.column_stack((speeds, price_per_mb))
```

```python

scatter = plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest

X_train, X_test = train_test_split(data, random_state=42)

clf = IsolationForest(max_samples=100, random_state=0)
clf.fit(X_train)

data = data[clf.predict(data) == 1]
```

```python
scatter = plt.scatter(data[:, 0], data[:, 1])
plt.show()
```
